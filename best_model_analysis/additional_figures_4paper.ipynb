```python
import numpy as np
import os
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import matplotlib.ticker as mticker
import matplotlib.gridspec as gridspec
from cartopy.mpl.ticker import (LongitudeFormatter, LatitudeFormatter)
import cartopy.util as cutil
from statsmodels.tsa.seasonal import seasonal_decompose
from scipy.stats import pearsonr
```

```python
years = np.arange(1980, 2022, 1) # from 1980 to 2021 included
project_dir = '/Users/huripari/Documents/PhD/TCs_Genesis'
data_dir = os.path.join(project_dir, 'data')
tcg_maps_dir = os.path.join(data_dir, 'IBTrACS', '2.5x2.5', 'TCG')
climate_indices_dir = os.path.join(data_dir, 'CI')
```

```python
# Load and compute total number of TCGs from 1980 to 2021
tcg_list = []
tcg_list_month = []
for y in years:
    file = os.path.join(tcg_maps_dir, 'TCG_2.5x2.5_{}.nc'.format(y))
    ds = xr.open_dataset(file)
    tcg_ts = ds.tcg.sum(dim=['latitude', 'longitude'])
    tcg = ds.tcg.sum(dim='time')
    tcg_list.append(tcg)
    tcg_list_month.append(tcg_ts)
tcg_year = xr.concat(tcg_list, dim='year')
tcg_tot = tcg_year.sum(dim='year')
tcg_month = xr.concat(tcg_list_month, dim='time')
# Load PDO index
filename = os.path.join(climate_indices_dir, 'PDO.txt')
date_range = pd.date_range(start=f'{1980}-01-01', end=f'{2021}-12-01', freq='MS')
df_pdo = pd.DataFrame(index=date_range, columns=['PDO'])
data = pd.read_table(filename, sep='\s+', header=None)
for r, row in enumerate(df_pdo.iterrows()):
    idx = df_pdo.index[r]
    month = idx.month
    year = idx.year
    df_pdo.loc[idx, 'PDO'] = data[(data[0] == year)][month].values[0]
```

```python
# Plot total number of tropical cyclones genesis from 1980 to 2021
# Make the data cyclic to avoid whit band at 360°E
var, clon, clat = cutil.add_cyclic(tcg_tot.values, tcg_tot.longitude.values, tcg_tot.latitude.values)
# Start figure and set coastlines
fig = plt.figure(figsize=(40, 8))
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(central_longitude=180))
ax.coastlines()
# Set gridlines, ticks and labels
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LongitudeFormatter()
gl.yformatter = LatitudeFormatter()
gl.xlocator = mticker.FixedLocator(np.arange(-180, 181, 20))
gl.ylocator = mticker.FixedLocator(np.arange(-90, 91, 10))
gl.xlabel_style = {'size': 18} 
gl.ylabel_style = {'size': 18}
# Plot the variable
field = ax.contourf(clon, clat, var, transform=ccrs.PlateCarree(), cmap='Reds')
cbar = plt.colorbar(field)
cbar.ax.tick_params(labelsize=18)

ax.set_title('Total number of tropical cyclones genesis from 1980 to 2021', size=26);
```

```python
# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(30, 6))

decomposition = seasonal_decompose(tcg_month.to_dataframe(), model='additive')
trend = decomposition.trend

trend_isnan = ~np.isnan(trend.values)
r, _ = pearsonr(df_pdo.values.flatten()[trend_isnan], trend.values[trend_isnan].flatten())

# Plot the first time series (PDO index) on the primary y-axis
color1 = 'tab:blue'
ax1.set_xlabel('Year', size=20)
ax1.set_ylabel('PDO', color=color1, size=20)
ax1.plot(df_pdo.index, df_pdo.values, color=color1, linewidth=2, label="PDO")  # Ensure proper x-values
ax1.tick_params(axis='y', labelcolor=color1, labelsize=18)
ax1.set_title(f'correlation: {r:.3f}', size=26)

# Create a second y-axis for the second time series (TCG)
ax2 = ax1.twinx()  
color2 = 'tab:red'
ax2.set_ylabel('Number of TCs', color=color2, size=20)
ax2.plot(tcg_month.time, trend.values, color=color2, linewidth=2, label="TCG")  # Ensure proper x-values
ax2.tick_params(axis='y', labelcolor=color2, labelsize=18)

# Set x-axis format properly
ax1.tick_params(axis='x', labelsize=16)
ax1.set_xlim([df_pdo.index.min(), df_pdo.index.max()])  # Set x-axis limits to match dataset

# Show legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
## Plot Time Series of final selected features overlayed with the number of TCs ##
BASIN = 'NWP'
# Dictionary of the basins and their corresponding cluster names
basin_dict = {'GLB': ('Global', 'nc12'), 'NEP': ('North East Pacific', 'DSnc12'), 'NWP': ('North West Pacific', 'Anc10'), 'NA': ('North Atlantic', 'DSnc6'),
              'NI': ('North Indian', 'DSnc12'), 'SI': ('South Indian', 'DSnc9'), 'SP': ('South Pacific', 'nc7')}
# Insert the selected featues for each basin in a dataframe
sel_features_list = []
for basin, (basin_name, cluster_data) in basin_dict.items():
    # Load the selected features
    sel_feat_perc_path = os.path.join(project_dir, 'FS_TCG', 'results', f'selected_features_best_models_{basin}_{cluster_data}.csv')
    df_sel_feat_perc = pd.read_csv(sel_feat_perc_path, index_col=0)
    selected_features = df_sel_feat_perc['80'].dropna().to_list()
    sel_features_list.append(selected_features)
# Create a dataframe with the selected features
df_sel_feature = pd.DataFrame([sel_feature for sel_feature in sel_features_list], index=basin_dict.keys())
# Load the file containing the time series of the predictors and of the target
nc_string = basin_dict[BASIN][1]
n_clusters = nc_string.split('nc')[1]
if 'DS' in nc_string:
    clusters_data = f'{BASIN}_{n_clusters}clusters_deseason'
elif 'A' in nc_string:
    clusters_data = f'{BASIN}_{n_clusters}clusters_anomaly'
else:
    clusters_data = f'{BASIN}_{n_clusters}clusters'
cluster_data_dir = os.path.join(project_dir, 'FS_TCG', 'data', f'{clusters_data}')
predictor_file = f'predictors_1970-2022_{n_clusters}clusters_8vars_9idxs.csv'
target_file = 'target_1970-2022_2.5x2.5.csv'
predictors_df = pd.read_csv(os.path.join(cluster_data_dir, predictor_file), index_col=0)
target_df = pd.read_csv(os.path.join(cluster_data_dir, target_file), index_col=0)
predictors_df.index = pd.to_datetime(predictors_df.index)
target_df.index = pd.to_datetime(target_df.index)
date_range = pd.date_range(start=f'{1980}-01-01', end=f'{2021}-12-01', freq='MS')
predictors_df = predictors_df[predictors_df.index.isin(date_range)]
target_df = target_df[target_df.index.isin(date_range)]
# List of final selected features
sel_features = df_sel_feature.loc[BASIN].dropna().to_list()
# Cycle through the selected features and plot the time series
for ff, feature in enumerate(sel_features):
    # Extract the time series of the feature
    feature_ts = predictors_df[feature]
    # Compute the correlation between the feature and the number of TCs
    r, _ = pearsonr(feature_ts.values.flatten(), target_df.values.flatten())
    # Create the figure and axis
    fig, ax1 = plt.subplots(figsize=(30, 6))
    # Plot the first time series (feature) on the primary y-axis
    color1 = 'tab:blue'
    ax1.set_xlabel('Year', size=20)
    ax1.set_ylabel(feature, color=color1, size=20)
    ax1.plot(feature_ts.index, feature_ts.values, color=color1, linewidth=2, label=feature)  # Ensure proper x-values
    ax1.tick_params(axis='y', labelcolor=color1, labelsize=18)
    ax1.set_title(f'correlation: {r:.3f}', size=26)
    # Create a second y-axis for the second time series (TCG)
    ax2 = ax1.twinx()
    color2 = 'tab:red'
    ax2.set_ylabel('Number of TCs', color=color2, size=20)
    ax2.plot(target_df.index, target_df.values, color=color2, linewidth=2, label="TCG")  # Ensure proper x-values
    ax2.tick_params(axis='y', labelcolor=color2, labelsize=18)
    # Set x-axis format properly
    ax1.tick_params(axis='x', labelsize=16)
    ax1.set_xlim([feature_ts.index.min(), feature_ts.index.max()])  # Set x-axis limits to match dataset
    # Show legend
    # ax1.legend(loc='upper left')
    # ax2.legend(loc='upper right')
    # Show the plot
    plt.show()
```

```python
## Plot time series of specific years coupled for comparison of values between them ##
# couple of years to compare for each basin
years_couples = {
    'GLB': ([1987, 1990], [1998, 1994], [2010, 2015])
    }
# most important features for each basin
important_features = {
    'GLB': ['sst_cluster9', 'abs_vo850_cluster5', 'mpi_cluster12']
}
metrics_units = {
    'GLB': ['°C', 's$^{-1}$', 'ms$^{-1}$']
}
# Plot the time series of each variable for the selected years
yy_couples = years_couples['GLB']
features_imp = important_features['GLB']
units = metrics_units['GLB']
for yy, y_cpl in enumerate(yy_couples):
    # Create the figure and axis
    fig = plt.figure(figsize=(18, 4))
    gs = gridspec.GridSpec(1, 3, figure=fig)
    for ff, feature in enumerate(features_imp):
        feature_ts_year1 = predictors_df.loc[predictors_df.index.year == yy_couples[yy][0], feature]
        feature_ts_year2 = predictors_df.loc[predictors_df.index.year == yy_couples[yy][1], feature]
        # Plot the time series of the feature for the 2 years
        ax1 = fig.add_subplot(gs[ff])
        ax1.set_xlabel('Month')
        ax1.set_ylabel(f'{units[ff]}', rotation='horizontal', ha='right')
        ax1.plot(feature_ts_year1.index.month, feature_ts_year1.values, label=f'{yy_couples[yy][0]} (min)', color='tab:blue', linewidth=2)
        ax1.plot(feature_ts_year2.index.month, feature_ts_year2.values, label=f'{yy_couples[yy][1]} (max)', color='tab:red', linewidth=2)
        ax1.tick_params(axis='y')
        ax1.set_title(feature)
        # Set x-axis format properly
        ax1.tick_params(axis='x')
        ax1.set_xlim([1, 12])
        # Show legend
        ax1.legend(loc='best')
        # Show the plot
    plt.show()
```

```python
mpi_list = []
sst_list = []
for y in years:
    mpi_file = os.path.join(data_dir, 'ERA5dv', '2.5x2.5', 'MPI', 'MPI_2.5x2.5_{}.nc'.format(y))
    sst_file = os.path.join(data_dir, 'ERA5', '2.5x2.5', 'MSLP-SST', 'ERA5_MSLP-SST_2.5x2.5_{}.nc'.format(y))
    ds = xr.open_dataset(mpi_file).mpi
    mpi_list.append(ds)
    ds = xr.open_dataset(sst_file).sst
    sst_list.append(ds)
mpi_ds = xr.concat(mpi_list, dim='time')
sst_ds = xr.concat(sst_list, dim='time')
```
